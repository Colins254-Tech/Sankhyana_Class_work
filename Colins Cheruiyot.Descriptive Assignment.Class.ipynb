Read the dataset “toys_dataset.csv” and find
(a) Mean, Median and Mode of Age
(b) Range, S.D. and variance of Age
(c) Plot the Normal distribution for random variable falling in the range
(-100, 100, 5).

```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
# Setting the working directory
os.chdir('E:/Sankhyana Consultancy/Mode 2 Assignments and Practice Exercises') 
# Reading the dataset
df = pd.read_csv('toy_dataset.csv')
# (a) Mean, Median, and Mode of Age
mean_age = df['Age'].mean()
median_age = df['Age'].median()
mode_age = df['Age'].mode()[0]  
print(f"Mean of Age: {mean_age}")
print(f"Median of Age: {median_age}")
print(f"Mode of Age: {mode_age}")

```

```python
# (b) Finding the Range, Standard Deviation, and Variance of Age
range_age = df['Age'].max() - df['Age'].min()
std_age = df['Age'].std()
variance_age = df['Age'].var()
print(f"Range of Age is: {range_age}")
print(f"Standard Deviation of Age is: {std_age}")
print(f"Variance of Age is: {variance_age}")
```

```python
# (c) Plotting the Normal distribution for random variable in range (-100, 100, 5)
x = np.arange(-100, 101, 5)
mean = 44.9502 
std_dev = 11.572485735982493
# Compute the probability density function (PDF) of the normal distribution
y = stats.norm.pdf(x, mean, std_dev)
# Plotting the distribution
plt.plot(x, y, label='Normal Distribution')
plt.title('Normal Distribution')
plt.xlabel('Random Variable')
plt.ylabel('Probability Density')
plt.grid(True)
plt.legend()
plt.show()
```

A candidate for public office has claimed that 60% of voters will vote for her. If 5
registered voters were sampled, what is the probability that exactly 3 would say they
favor this candidate? Write a code for this.

```python
from scipy.stats import binom
# Parameters
n = 5          
p = 0.60        
k = 3           
# Calculate the probability
probability = binom.pmf(k, n, p)
print(f"The probability that exactly 3 out of 5 voters favor the candidate is: {probability:.4f}")
```

Rakesh flips a fair coin 6 times. What is the probability that the coin lands on heads 3
times or fewer? Write a code for binomial distribution to find the probability of getting
head.

```python
from scipy.stats import binom
# Parameters
n = 6         
p = 0.5       
k = 3      
# Calculating the cumulative probability of getting 3 or fewer heads
probability = binom.cdf(k, n, p)
print(f"The probability of getting heads 3 times or fewer in 6 flips is: {probability:.4f}")
```

(M2C02e2.04)
The weight of a certain species of frog is uniformly distributed between 15 and 25
grams. If you randomly select a frog, what is the probability that the frog weighs
between 17 and 19 grams? Write a code to plot the uniform distribution.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import uniform
# Parameters
a = 15  # Minimum weight
b = 25  # Maximum weight
# Uniform distribution setup
distribution = uniform(loc=a, scale=b-a)
# Calculating the probability of weight being between 17 and 19 grams
probability = distribution.cdf(19) - distribution.cdf(17)
print(f"The probability that the frog weighs between 17 and 19 grams is: {probability:.4f}")
# Plotting the uniform distribution
x = np.linspace(a-1, b+1, 1000)  # Generating values for the x-axis
y = distribution.pdf(x)  # Probability density function for uniform distribution
plt.plot(x, y, 'b-', label='Uniform Distribution')
plt.fill_between(x, y, where=((x >= 17) & (x <= 19)), color='lightblue', label='17 to 19 grams')
plt.title('Uniform Distribution of Frog Weights')
plt.xlabel('Weight (grams)')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()
```

M2C02e2.05
The length of an NBA game is uniformly distributed between 120 and 170 minutes.
What is the probability that a randomly selected NBA game lasts more than 150
minutes? Write a code for this

```python
from scipy.stats import uniform
# Parameters
a = 120 
b = 170
x = 150 
# Uniform distribution setup
distribution = uniform(loc=a, scale=b-a)
# Calculate the probability that the game lasts more than 150 minutes
probability = 1 - distribution.cdf(x)
print(f"The probability that an NBA game lasts more than 150 minutes is: {probability:.4f}")
```
